## Cleaning Data

### Deleting rows with NA values and ID column

```python
# Deleting rows with missing values 
credit = credit.dropna(axis=0)
credit.drop('ID', axis=1)
```

### Replacing values in variables

```python
# Converting PAY_* -2 to 10 and -1 to 11
credit["PAY_0"].replace({-1 : 10, -2 : 11}, inplace=True)
credit["PAY_2"].replace({-1 : 10, -2 : 11}, inplace=True)
credit["PAY_3"].replace({-1 : 10, -2 : 11}, inplace=True)
credit["PAY_4"].replace({-1 : 10, -2 : 11}, inplace=True)
credit["PAY_5"].replace({-1 : 10, -2 : 11}, inplace=True)
credit["PAY_6"].replace({-1 : 10, -2 : 11}, inplace=True)

# Converting Other Education in the same value (0) 
credit["EDUCATION"].replace({4 : 0, 5 : 0, 6 : 0}, inplace=True)
```

## Correlation Matrices 

### Correlation Matrix using the default variable types (Numerical int64)

```python
# Correlation Matrix with no categorical attributes 
corrMatrix = credit.corr()
print(corrMatrix)
```

### Getting the top correlated pairs and correlation values with dependent variable

```python
# Getting the top 10 correlated pairs 

# Get top 25 correlation pairs
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top Absolute Correlations")
print(get_top_abs_correlations(credit, 10))
print()

print("Absolute Correlations with Y variable")
corrY = credit.corr()['DEFAULT'][:]
print(corrY)
```

```python
# Correlation Matrix using Dummy variables
credit_dummies = credit

credit_dummies["SEX"].replace({1 : "SEX_MALE", 2 : "SEX_FEMALE"}, inplace=True)
credit_SEX_dummies = pd.get_dummies(credit_dummies['SEX'])

credit_dummies["EDUCATION"].replace({0 : "EDUC_OTHER", 1 : "EDUC_GRAD", 2 : "EDUC_UNIV", 3 : "EDUC_HS"}, inplace=True)
credit_EDUCATION_dummies = pd.get_dummies(credit_dummies['EDUCATION'])

credit_dummies["MARRIAGE"].replace({0 : "MARR_OTHER", 1 : "MARR_MARRIED", 2 : "MARR_SINGLE", 3 : "MARR_DIVORCED"}, inplace=True)
credit_MARRIAGE_dummies = pd.get_dummies(credit_dummies['MARRIAGE'])

credit_dummies["PAY_0"].replace({0 : "PAY_0_PAIDINFULL", 1 : "PAY_0_REVCREDIT", 2 : "PAY_0_2_MONTH", 3 : "PAY_0_3_MONTH", 4 : "PAY_0_4_MONTH", 5 : "PAY_0_5_MONTH", 6 : "PAY_0_6_MONTH", 7 : "PAY_0_7_MONTH", 8 : "PAY_0_8_MONTH", 9 : "PAY_0_9PLUS_MONTH", 10 : "PAY_0_NOCONS", 11 : "PAY_0_PAIDFULL"}, inplace=True)
credit_PAY_0_dummies = pd.get_dummies(credit_dummies['PAY_0'])

credit_dummies["PAY_2"].replace({0 : "PAY_2_PAIDINFULL", 1 : "PAY_2_REVCREDIT", 2 : "PAY_2_2_MONTH", 3 : "PAY_2_3_MONTH", 4 : "PAY_2_4_MONTH", 5 : "PAY_2_5_MONTH", 6 : "PAY_2_6_MONTH", 7 : "PAY_2_7_MONTH", 8 : "PAY_2_8_MONTH", 9 : "PAY_2_9PLUS_MONTH", 10 : "PAY_2_NOCONS", 11 : "PAY_2_PAIDFULL"}, inplace=True)
credit_PAY_2_dummies = pd.get_dummies(credit_dummies['PAY_2'])

credit_dummies["PAY_3"].replace({0 : "PAY_3_PAIDINFULL", 1 : "PAY_3_REVCREDIT", 2 : "PAY_3_2_MONTH", 3 : "PAY_3_3_MONTH", 4 : "PAY_3_4_MONTH", 5 : "PAY_3_5_MONTH", 6 : "PAY_3_6_MONTH", 7 : "PAY_3_7_MONTH", 8 : "PAY_3_8_MONTH", 9 : "PAY_3_9PLUS_MONTH", 10 : "PAY_3_NOCONS", 11 : "PAY_3_PAIDFULL"}, inplace=True)
credit_PAY_3_dummies = pd.get_dummies(credit_dummies['PAY_3'])

credit_dummies["PAY_4"].replace({0 : "PAY_4_PAIDINFULL", 1 : "PAY_4_REVCREDIT", 2 : "PAY_4_2_MONTH", 3 : "PAY_4_3_MONTH", 4 : "PAY_4_4_MONTH", 5 : "PAY_4_5_MONTH", 6 : "PAY_4_6_MONTH", 7 : "PAY_4_7_MONTH", 8 : "PAY_4_8_MONTH", 9 : "PAY_4_9PLUS_MONTH", 10 : "PAY_4_NOCONS", 11 : "PAY_4_PAIDFULL"}, inplace=True)
credit_PAY_4_dummies = pd.get_dummies(credit_dummies['PAY_4'])

credit_dummies["PAY_5"].replace({0 : "PAY_5_PAIDINFULL", 1 : "PAY_5_REVCREDIT", 2 : "PAY_5_2_MONTH", 3 : "PAY_5_3_MONTH", 4 : "PAY_5_4_MONTH", 5 : "PAY_5_5_MONTH", 6 : "PAY_5_6_MONTH", 7 : "PAY_5_7_MONTH", 8 : "PAY_5_8_MONTH", 9 : "PAY_5_9PLUS_MONTH", 10 : "PAY_5_NOCONS", 11 : "PAY_5_PAIDFULL"}, inplace=True)
credit_PAY_5_dummies = pd.get_dummies(credit_dummies['PAY_5'])

credit_dummies["PAY_6"].replace({0 : "PAY_6_PAIDINFULL", 1 : "PAY_6_REVCREDIT", 2 : "PAY_6_2_MONTH", 3 : "PAY_6_3_MONTH", 4 : "PAY_6_4_MONTH", 5 : "PAY_6_5_MONTH", 6 : "PAY_6_6_MONTH", 7 : "PAY_6_7_MONTH", 8 : "PAY_6_8_MONTH", 9 : "PAY_6_9PLUS_MONTH", 10 : "PAY_6_NOCONS", 11 : "PAY_6_PAIDFULL"}, inplace=True)
credit_PAY_6_dummies = pd.get_dummies(credit_dummies['PAY_6'])

credit_dummies = pd.concat([credit_dummies, credit_SEX_dummies, credit_EDUCATION_dummies, credit_MARRIAGE_dummies, credit_PAY_0_dummies, credit_PAY_2_dummies, credit_PAY_3_dummies, credit_PAY_4_dummies, credit_PAY_5_dummies, credit_PAY_6_dummies], axis=1)
del credit_dummies['SEX']
del credit_dummies['EDUCATION']
del credit_dummies['MARRIAGE']
del credit_dummies['PAY_0']
del credit_dummies['PAY_2']
del credit_dummies['PAY_3']
del credit_dummies['PAY_4']
del credit_dummies['PAY_5']
del credit_dummies['PAY_6']

#Delete ID column 
del credit_dummies['ID']

credit_dummies.head()
```
